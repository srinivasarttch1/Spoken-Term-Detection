```python
import os
from sphfile import SPHFile
#from pydub import AudioSegment
import numpy as np
train='TRAIN'
dialects=[os.path.join(train,p) for p in sorted(os.listdir(train))]
speakers=[]
for i in range(len(dialects)):
    speakers.append([os.path.join(dialects[i],p) for p in sorted(os.listdir(dialects[i]))])#list of lists(speakers of one dialect)
sa1_word_path=[]
sa1_wav_path=[]
sa2_word_path=[]
sa2_wav_path=[]
sa1_text_path=[]
sa2_text_path=[]
i=0
j=0
for i in range(len(speakers)):
    for j in range(len(speakers[i])):
        sa1_wav_path.append(os.path.join(speakers[i][j],'SA1.WAV'))
        sa2_wav_path.append(os.path.join(speakers[i][j],'SA2.WAV'))
        sa1_word_path.append(os.path.join(speakers[i][j],'SA1.WRD'))
        sa2_word_path.append(os.path.join(speakers[i][j],'SA2.WRD'))
        sa1_text_path.append(os.path.join(speakers[i][j],'SA1.txt'))
        sa2_text_path.append(os.path.join(speakers[i][j],'SA2.txt'))
```

```python
len(sa1_wav_path)
```

```python
#for sa1, words seperation
for i in range(len(sa1_wav_path)):
    split=sa1_wav_path[i].split("\\")
    sph =SPHFile(
        sa1_wav_path[i]
    )
    speaker_name=split[2]
    f = open(sa1_word_path[i], "r")
    for line in f:
        t=line.split(" ")
        start_time=t[0]
        end_time=t[1]
        word_name=t[2]
        output_file_name=speaker_name+"_"+word_name
        output_file=output_file_name.split("\n")[0]+'.WAV'
        #print(output_file[0])
        #print(word_name.split("\\")[0])
        output_path=os.path.join('train_words',word_name.split("\n")[0])
        output_path=os.path.join(output_path,output_file)
        sph.write_wav( output_path,int(start_time)/16000, int(end_time)/16000 )

    
  
```

```python
#for sa2 words seperation
for i in range(len(sa2_wav_path)):
    split=sa2_wav_path[i].split("\\")
    sph =SPHFile(
        sa2_wav_path[i]
    )
    speaker_name=split[2]
    f = open(sa2_word_path[i], "r")
    for line in f:
        t=line.split(" ")
        start_time=t[0]
        end_time=t[1]
        word_name=t[2]
        output_file_name=speaker_name+"_"+word_name
        output_file=output_file_name.split("\n")[0]
        output_file=output_file+".WAV"
        #print(output_file[0])
        #print(word_name.split("\\")[0])
        output_path=os.path.join('train_words',word_name.split("\n")[0])
        output_path=os.path.join(output_path,output_file)
        sph.write_wav( output_path,int(start_time)/16000, int(end_time)/16000 )

    
  
```
